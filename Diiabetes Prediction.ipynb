```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn import svm
from sklearn.metrics import accuracy_score
```

```python
# Loading the data dataset in a pandas dataframe :
path = "/content/drive/MyDrive/dataset_1/Diabetes_data.csv"
diabetes_data = pd.read_csv(path)
```

```python
diabetes_data.groupby('Outcome').mean()
```

```python
# Separating data and Labels :

X = diabetes_data.drop(columns='Outcome', axis =1)
Y = diabetes_data['Outcome']
```

```python
print(X)
print(Y)
```

```python
# Train test split :
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.2, stratify=Y, random_state=2)
print(X.shape, X_train.shape, X_test.shape)
```

```python
classifier =svm.SVC(kernel='linear')
classifier.fit(X_train, Y_train)
```

```python
# Accuracy score on training data :
X_train_prediction = classifier.predict(X_train)
training_data_accuracy = accuracy_score(X_train_prediction, Y_train)
print('Accuracy of training data :', training_data_accuracy)
```

```python
# Accuracy score on test data :
X_test_prediction = classifier.predict(X_test)
test_data_accuracy = accuracy_score(X_test_prediction, Y_test)
print('Accuracy of test data :', test_data_accuracy)
```

<b> Making a Predictive System : </b>

```python
input_data = (4,110,92,0,0,37.6,0.191,30)

# Changing the data to a numpy array :
new_input = np.asarray(input_data)

# Standadize the input data :# Reshape the array as we are predicting for one instance :
reshaped_input = new_input.reshape(1,-1)
prediction = classifier.predict(reshaped_input)
print(reshaped_input)
print(prediction)
```
